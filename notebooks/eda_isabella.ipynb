# EDA de Isabella – Titanic

Exploratory Data Analysis (EDA) sobre el dataset de pasajeros del Titanic.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (8, 4)

df = pd.read_csv('data/raw/titanic.csv')
df.head()
```

## 1. Información general del dataset


```python
df.info()
```

## 2. Descripción estadística de las variables numéricas


```python
df.describe()
```

## 3. Valores nulos por columna


```python
df.isna().sum()
```

## 4. Distribución de la edad


```python
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title('Distribución de la edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

## 5. Tasa de supervivencia global


```python
df['survived'].value_counts(normalize=True)
```

## 6. Supervivencia por sexo


```python
sns.barplot(x='sex', y='survived', data=df)
plt.title('Proporción de supervivencia por sexo')
plt.show()
```
