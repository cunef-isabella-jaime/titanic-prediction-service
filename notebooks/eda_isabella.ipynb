## 4. Distribuci√≥n de la edad


```python
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title('Distribuci√≥n de la edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

La mayor√≠a de los pasajeros tiene entre 20 y 40 a√±os.

Hay presencia de ni√±os peque√±os, lo cual puede ser relevante ya que hist√≥ricamente se prioriz√≥ salvar a mujeres y ni√±os.

Se observan valores faltantes en la columna Age, lo cual confirma la necesidad de imputaci√≥n en la fase de preprocesamiento.

## 5. Tasa de supervivencia global


```python
df['survived'].value_counts(normalize=True)
```

El porcentaje de supervivientes permite observar si el dataset est√° balanceado.
Generalmente, alrededor del 38% de los pasajeros sobrevivi√≥, lo que indica un desbalanceo moderado entre clases.

## 6. Supervivencia por sexo


```python
sns.barplot(x='sex', y='survived', data=df)
plt.title('Proporci√≥n de supervivencia por sexo')
plt.show()
```

Interpretaci√≥n ‚Äî Supervivencia seg√∫n sexo

Las mujeres presentan una tasa de supervivencia significativamente mayor.

Esto confirma la pol√≠tica de evacuaci√≥n "women and children first".

La variable Sex ser√° probablemente una de las m√°s importantes para los modelos de ML.
